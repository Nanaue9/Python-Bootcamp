
# Getting Started with Google Colaboratory (Colab)

In this exercise, we will get familiar with Google Colaboratory (Colab), an online platform for writing and running Python code. Follow these steps to complete the exercise and learn basic operations in Colab.

## Step 1: Open Google Colab
Visit [Google Colab](https://colab.research.google.com/) and sign in with your BU account.

## Step 2: Create a New Notebook
Click on "File" -> "New Notebook" to create a new Colab notebook.

## Step 3: Add a Markdown Cell
Click on "+ Text" to add a new Markdown cell. Write a description similar to the following and press `Shift + Enter` to render it:

```markdown
# Python Version
The following code provides the version of the Python that this notebook is running on:
```

## Step 4: Add a Code Cell to Display Python Version
Click on "+ Code" to add a new code cell and type the following code. Run the cell by pressing `Shift +

 Enter` to display the Python version.


````python
import sys
print(sys.version)
````


## Step 5: Rename the Notebook
Click on the notebook name at the top left and give it a meaningful name, such as "Exercise 1 - Colab Introduction".

## Step 6: Install and Import a Library
Add a new code cell and write code to install a new library using `pip`. After installing, import the library and display its version:

```python
!pip install numpy
import numpy as np
print("NumPy version:", np.__version__)
```

## Step 7: Save the Notebook to Google Drive
Click on "File" -> "Save a Copy in Drive" to save your notebook in your Google Drive for future reference.
